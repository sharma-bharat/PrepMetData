```python
import pandas as pd
import requests

# Define the URL
url = 'https://wrcc.dri.edu/cgi-bin/wea_daysum2.pl?stn=nduk&day=01&mon=01&yea=23&unit=M'

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Split the response content into lines
    lines = response.text.splitlines()
    
    # Assuming the data of interest starts at line 26 (adjust as needed)
    data_lines = lines[26:]
    
    # Parse the data into a pandas DataFrame
    df = pd.DataFrame([line.split() for line in data_lines])#, columns=["Date", "Variable1", "Variable2", ...])  # Add column names
    
    # Display the DataFrame
    print(df)
else:
    print("Failed to retrieve data from the URL.")
```
